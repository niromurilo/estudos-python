### Puxar base de dados

```python
import pandas as pd
base_dados_df = pd.read_csv(r"exportacoes_franca.csv")
display(base_dados_df)
```

### Como foi a evolução das exportações para a frança ao longo dos anos?

Crie uma tabela com os anos de 2016, 2017, 2018, 2019 e 2020 e o valor US$ de exportações

```python
base_anos_valor = base_dados_df[["Year", "US$ FOB"]].groupby("Year").sum()
def formatar_numero(valor):
    valor_formatado = f"US${valor:,}"
    return valor_formatado
base_anos_valor["US$ FOB"] = base_anos_valor["US$ FOB"].apply(formatar_numero)
display(base_anos_valor)
```

### Quais os produtos mais exportados ao longo de todo o período?

Queremos saber os produtos com maior valor US$ de exportação

```python
base_produtos_mais_export = base_dados_df[["SH2 Description", "US$ FOB"]].groupby("SH2 Description").sum()
base_produtos_mais_export = base_produtos_mais_export.sort_values(by="US$ FOB", ascending=False)

#formatar sempre por ultimo
base_produtos_mais_export["US$ FOB"] = base_produtos_mais_export["US$ FOB"].apply(formatar_numero)
display(base_produtos_mais_export)
```

### Em 2020 qual cidade mais exportou para a França?
Em valores US$

```python
base_2020 = base_dados_df.loc[base_dados_df["Year"]==2020, :]

base_2020_city = base_2020[["City", "US$ FOB"]].groupby("City").sum()
base_2020_city = base_2020_city.sort_values(by="US$ FOB", ascending=False)
base_2020_city["US$ FOB"] = base_2020_city["US$ FOB"].apply(formatar_numero)
display(base_2020_city)
```

### Quais os produtos mais exportados (em US$) que as 2 maiores cidades (em exportação em 2020) exportaram?

```python
cidade = base_2020_city.index[0]

base_dados_cidade = base_2020.loc[base_2020["City"]==cidade]

base_produtos_cidade = base_dados_cidade[["SH2 Description", "US$ FOB"]].groupby("SH2 Description").sum()

base_produtos_cidade = base_produtos_cidade.sort_values(by="US$ FOB", ascending=False)

base_produtos_cidade["US$ FOB"] = base_produtos_cidade["US$ FOB"].apply(formatar_numero)

print(f"Produtos mais exportados da Cidade: {cidade}")
display(base_produtos_cidade)
```
